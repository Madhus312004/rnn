```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'The glass of milk',
    'The glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
voc_size=10000
```

```python
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr  
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_len=8
```

```python
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)
embedded_docs
```

```python
#dimentsion features
dim=10
```

```python
model_ex=Sequential()
model_ex.add(Embedding(voc_size,dim,input_length=sent_len))
model_ex.compile('adam','mse')
```

```python
model_ex.summary()
```

```python
model_ex.predict(embedded_docs) #all sentance of var into 10 vector values for each word in sentance.
```

```python
model_ex.predict(embedded_docs[0]) #one sentance 8 words into each word 10 vectors so 80 vector for a single sentance
```
